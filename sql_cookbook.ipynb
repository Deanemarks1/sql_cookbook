```python


# DB SETUP OCT 22 - 2025
#-----------------------------------------------------------------------------------------------------------
import requests

# list of files to pull from your repo
files_to_pull = [
    "mysqlconnector.py",
    "git_cookbook.ipynb"
]

base_url = "https://raw.githubusercontent.com/Deanemarks1/sql_cookbook/main/"

for file_name in files_to_pull:
    url = base_url + file_name
    print(f"ðŸ“¥ Pulling {file_name} from GitHub...")
    
    r = requests.get(url)
    r.raise_for_status()

    with open(file_name, "wb") as f:
        f.write(r.content)
        print(f"âœ… Overwrote local copy: {file_name}")

    # execute only .py files automatically
    if file_name.endswith(".py"):
        with open(file_name) as f:
            code = compile(f.read(), file_name, "exec")
            exec(code, globals())
            print(f"âš¡ Executed: {file_name}")

print("ðŸ”¥ All GitHub files updated.")
#-----------------------------------------------------------------------------------------------------------

```
